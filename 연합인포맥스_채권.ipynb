```
import pandas as pd
import numpy as np
import urllib.request 
import bs4
```

```

## 기사별 url 정보를 수집한다. 
title_list = []
link_list = []
date_list = []

for i in range(1,10):
    
    print(i)
    i = str(i)
    URL = 'https://news.einfomax.co.kr/news/articleList.html?page={}'.format(i) +'&total=86231&box_idxno=&sc_area=A&view_type=sm&sc_word=%EC%B1%84%EA%B6%8C'

    html = urllib.request.urlopen(URL)
    bs_obj = bs4.BeautifulSoup(html, "html.parser")

    title_txt = bs_obj.findAll('div', {'class': 'list-titles'})
    body_txt = bs_obj.findAll('p', {'class': 'list-summary'})
    date_txt = bs_obj.findAll('div',  {'class' : 'list-dated'})

    for title in title_txt:
        title_name = title.find('a').get_text()
        title_list.append(title_name)

    for dt in date_txt:
        dt_tag = dt.get_text()
        date_list.append(dt_tag)

    for body in body_txt:
        atag = body.find('a')
        link = atag['href']
        print(link)
        address = 'https://news.einfomax.co.kr/'
        new_link = address + link
        link_list.append(new_link)

    df = pd.DataFrame([title_list, link_list, date_list]).T
    df.columns = ['title', 'url', 'cat_date']
    df = df.iloc[:-1]

#    df['category'] = df['cat_date'].apply(lambda x: str(x).split('|'))
#    for list in df['category']:
#        df['date'] = list[2]
#        df['date'] = df['date'].apply(lambda x: str(x).split(' ')[1])
```

```
df['category'] = df['cat_date'].apply(lambda x: str(x).split('|'))

dt_list = []
cat_name = []
for i in range(len(df)):
    tmp = df.iloc[i]['category']
    dt = tmp[-1]
    cat = tmp[0]
    dt_list.append(dt)
    cat_name.append(cat)

df = pd.concat([df, pd.Series(dt_list), pd.Series(cat_name)], axis=1)
df['Date'] = df[0].apply(lambda x: str(x)[1:11])
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
```

```
df.columns = ['title', 'url', 'cat_date', 'category_and_date', 'date', 'category', 'Date']
df = df[['Date','title', 'url', 'category']]
df
```

```
content_list = []

for i in range(50):
    print(i)
    txt_url = df['url'].values.tolist()
    txt_url = txt_url[i]
    print(txt_url)
    html = urllib.request.urlopen(txt_url)
    bs_obj = bs4.BeautifulSoup(html, "html.parser")
    
    try:
        content_txt = bs_obj.find('div', {'id': 'article-view-content-div'}).get_text()
        print(content_txt)
        content_txt = content_txt.replace('\n', '')
        content_txt = content_txt.replace('\r', '')
        content_txt = content_txt.replace('\t', '')
        content_list.append((txt_url, content_txt))
    except:
      pass
```

```
cont_df = pd.DataFrame(content_list)
cont_df.columns = ['url', 'text']

report_df = pd.merge(df, cont_df, on='url', how='inner')
```
